## requests 簡介

在爬蟲中，我們會使用 requests 把原始資料從網頁伺服器 (Web Server) 中讀取回來。

* 使用 get 方法讀取網站原始碼

```python
r = requests.get(url)
```

* 使用 post 方法讀取網站原始碼

```python
r = requests.post(url, data=data)
```

參考連結：[Python Requests.Quickstart](http://docs.python-requests.org/en/master/user/quickstart/)
