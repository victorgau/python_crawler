# BeautifulSoup的select語法說明

```python
from bs4 import BeautifulSoup
```

```python
s1 = """
<body>

<h1>哈囉！HTML！</h1>

<div class="class1">
    <p>這是第一個p標籤內容！</p>
    <p>這是第二個p標籤內容！</p>
</div>

<div class="class2">
    <p>這是第三個p標籤內容！</p>
    <p>這是第四個p標籤內容！</p>
</div>
</body>
"""
```

```python
soup1 = BeautifulSoup(s1)
```

```python
soup1.select('div.class1 p')
```

---

```python
s2 = """
<body>

<h1>哈囉！HTML！</h1>

<div class="class1">
    <p>這是第一個p標籤內容！</p>
    <div>
        <p>這是第二個p標籤內容！</p>
    </div>
</div>

<div class="class2">
    <p>這是第三個p標籤內容！</p>
    <p>這是第四個p標籤內容！</p>
</div>
</body>
"""
```

```python
soup2 = BeautifulSoup(s2)
```

```python
soup2.select("div.class1 p")
```

```python
soup2.select("div.class1 > p")
```

```python
soup2.select("div.class1 div p")
```

---

```python
s3 = """
<body>

<h1>哈囉！HTML！</h1>

<div class="class1">
    <p>這是第一個p標籤內容！</p>
    <div>
        <p data-no="two">這是第二個p標籤內容！</p>
    </div>
</div>

<p class="class1">這是第三個p標籤內容！</p>
<p id="p4">這是第四個p標籤內容！</p>
</body>
"""
```

```python
soup3 = BeautifulSoup(s3)
```

```python
soup3.select(".class1")
```

```python
soup3.select("#p4")
```

```python
soup3.select('p[data-no="two"]')
```
